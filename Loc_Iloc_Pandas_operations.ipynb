The iloc indexer syntax is data.iloc[row selection, column selection]. “iloc” in pandas is used to select rows and columns by number. 
Used to select the rows and columns by their number.









# Selecting Rows using iloc

```
print(data.iloc[0]) # first row of data frame 
print('\n\n')
print(data.iloc[1]) # second row of data frame 
print('\n\n')
print(data.iloc[-1]) # Last row of data frame
```

# Selecting Columns Using Iloc.

```
print(data.iloc[:,0]) # first column of data frame (first_name)
print('\n\n')
print(data.iloc[:,1]) # second column of data frame (last_name)
print('\n\n')
print(data.iloc[:,-1]) # last column of data frame (id)
```

# Selecting Multiple Rows and Columns of a Dataframe suing iloc.

```
print(data.iloc[0:5]) # first five rows of dataframe
print('\n\n')
print(data.iloc[:, 0:2]) # first two columns of data frame with all rows
print('\n\n')
print(data.iloc[[0,3,6,24], [0,5,6]]) # 1st, 4th, 7th, 25th row + 1st 6th 7th columns.
print('\n\n')
print(data.iloc[0:5, 5:8]) # first 5 rows and 5th, 6th, 7th columns of data frame (county -> phone1)
```

# Datatypes Using Iloc
.iloc returns a Pandas Series when one row is selected, and a Pandas DataFrame when multiple rows are selected, or if any column in full is selected

```
print(type(data.iloc[100])) # Result is Series because only one row is selected.
print(type(data.iloc[[100]])) # Result is of type Dataframe because we use list selection.
print(type(data.iloc[1:3,5])) # Result is Series because single column is selected.
print(type(data.iloc[1:3,[5]])) # Result is Dataframe because we use list selection in DataFrame.
print(type(data.iloc[1:3,3:6])) # Dataframe because Multiple Rows and columns selected.
```



# Loc Selection
The Pandas loc indexer can be used with DataFrames for two different use cases:

1.   Selecting rows by label/index
2.   Selecting rows with a boolean / conditional lookup.




---

Syntax is same as iloc.
data.loc[row selection, column selection]. 




```
# Inplace :- Modify the DataFrame in place (do not create a new object).
# Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length).
# The index can replace the existing index or expand on it.
# Example :- Suppose we have an id column and we want that id column to behave as our new index we can do it by using set_index function.
data.set_index("last_name", inplace=True)
data.head()
```

```
print(data.loc['Andrade'])
# Select rows with index values 'Andrade' and 'Veness', with all columns between 'city' and 'email'
print('\n\n')
print(data.loc[['Andrade', 'Veness'], 'city':'email'])
# # Select same rows, with just 'first_name', 'address' and 'city' columns
print('\n\n')
print(data.loc['Andrade':'Veness', ['first_name', 'address', 'city']])
print('\n\n')
 
```

```
# Change the index to be based on the 'id' column
data.set_index('id', inplace=True)
print('\n\n')
```

```
# select the row with 'id' = 11
print(data.head())
print(data.loc[11])
print(data.iloc[11])
# df.reset_index(drop=True, inplace=True)
```

```
# Select rows with first name Antonio, # and all columns between 'city' and 'email'
print(data.loc[data['first_name'] == 'Antonio', 'city':'email'])
print('\n\n') 
# Select rows where the email column ends with 'hotmail.com', include all columns
print(data.loc[data['email'].str.endswith("hotmail.com")])   
print('\n\n')
# Select rows with last_name equal to some values, all columns
print(data.loc[data['last_name'].isin(['jong ung', 'Tyisha', 'Eric'])])   
print('\n\n')       
# Select rows with first name Antonio AND gmail email addresses
print(data.loc[data['email'].str.endswith("gmail.com") & (data['first_name'] == 'Antonio')]) 
print('\n\n') 



```

# Merging Dataframes.


```
# We are reading Three Data frames. sep=','
# Delimiter to use. If sep is None, the C engine cannot automatically detect the separator, but the Python parsing engine can,
# meaning the latter will be used and automatically detect the separator by Python’s builtin sniffer tool, csv.Sniffer.
user_usage = pd.read_csv('https://raw.githubusercontent.com/Priyankk18k/Data_sets/master/user_usage.csv')
user_devices = pd.read_csv('https://raw.githubusercontent.com/Priyankk18k/Data_sets/master/user_device.csv')
android_devices = pd.read_csv('https://raw.githubusercontent.com/Priyankk18k/Data_sets/master/android_devices.csv')
```

```
user_usage.head()
```

```
user_devices.head()
```

```
android_devices.head(6)
```

```
# By default, the Pandas merge operation acts with an “inner” merge. 
#An inner merge, (or inner join) keeps only the common values in both the left and right dataframes for the result.
result_df = pd.merge(user_usage,user_devices[['use_id', 'platform', 'device']], on='use_id')
```

```
result_df
```

# Merging without specifying The columns.

```
result_new_df = pd.merge(user_usage, user_devices, on='use_id')
result_new_df
```

```
user_usage['use_id'].isin(user_devices['use_id']).value_counts()
```

# Different Merge Types Available.

1.   Inner Merge/join
2.   Left Merge/join

3.   Right Merge/join
4.   Full outer Merge/join






```
# For right i will put how='right and for outer how = 'outer'.
result_left_df = pd.merge(user_usage, user_devices, on='use_id', how='left')
result_left_df
```

```
result_outer_df = pd.merge(user_usage, user_devices, on='use_id', how='outer', indicator= True)
result_outer_df
# Both means data is from both Dataframe, Right means from right Df , Left Means from left Dataframe.
# To Specify coulmns of left and right dataframe you can right left_on and right_on parameters.
```
